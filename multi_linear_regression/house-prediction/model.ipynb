```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/house_price_regression_dataset.csv')

```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split


# Sp;itting the data into features and target variable

# Feature
x = df.drop('House_Price', axis=1)

# Target variable(Label)
y = df['House_Price']

# using the train test split function
X_train, X_test, y_train, y_test = train_test_split(x,y,
                                   test_size=0.20, 
                                   random_state=101,
                                   shuffle=True)

# printing out train and test sets

print('X_train : ')
print(X_train.size)
print('')
print('X_test : ')
print(X_test.size)
print('')
print('y_train : ')
print(y_train.size)
print('')
print('y_test : ')
print(y_test.size)
```

## Dataset Splitting:

the [kaggle house price regression dataset](../data/house_price_regression_dataset.csv), is made up of 10000 rows and 8 columns, of which one column is the house price column(dependent variable Y), we use `train_test_split` from sklearn to split it into 4 dataframes:
X_train
y_train

X_test
y_test

A training set and a test set, where the split is 80/20 respectively. The training set we feed into the model containing both the independent variables and dependent ones in order to train it, and then we predict with the X_test(Testing dataframe)

```python
model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)


plt.scatter(y_pred, y_test, alpha=0.4)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')  # perfect line
plt.xlabel("Predicted Price")
plt.ylabel("Actual Price")
plt.title("Predicted vs Actual")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error: {mae}")
print(f"Root Mean Squared Error: {rmse}")
print(f"R^2 Score: {r2}")
```
